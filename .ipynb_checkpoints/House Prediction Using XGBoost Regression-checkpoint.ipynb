```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn.datasets 
from sklearn.model_selection import train_test_split
from sklearn import metrics
from xgboost import XGBRegressor
```

```python
house_price_dataset = sklearn.datasets.load_boston()
```

```python
house_price_dataset_dataframe= pd.DataFrame(house_price_dataset.data,columns=house_price_dataset.feature_names)
```

```python
house_price_dataset_dataframe.head()
```

```python
house_price_dataset_dataframe['PRICE']= house_price_dataset.target
```

```python
house_price_dataset_dataframe
```

```python
house_price_dataset_dataframe.shape
```

```python
house_price_dataset_dataframe.isnull().sum()
```

```python
house_price_dataset_dataframe.describe()
```

```python
house_price_dataset_dataframe.info()
```

```python
correlation = house_price_dataset_dataframe.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(data=correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```

```python
X= house_price_dataset_dataframe.drop(columns='PRICE',axis=1) #Features
Y=house_price_dataset_dataframe['PRICE'] #Labels
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
X_test.shape
```

```python
X_train.shape
```

```python
model = XGBRegressor()
model.fit(X_train,Y_train)
```

```python
predictor= model.predict(X_train)
predictor
```

```python
# R squared Error

score_1= metrics.r2_score(Y_test,predictor)
score_1
```

```python
#mean absolute error
score_2= metrics.mean_absolute_error(Y_test,predictor)
```

```python
score_2
```

```python
plt.scatter(Y_test,predictor)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Price vs Predicted Price')
```
